3. Dropping unwanted columns

   DATA REDUCTION

```python
mean_Data_Value = dataset1['Data_Value'].mean()
dataset1['Data_Value'] = dataset1['Data_Value'].fillna(mean_Data_Value)
```

```python
dataset1.info()
```

```python
mean_Data_Value_Alt = dataset1['Data_Value_Alt'].mean()
dataset1['Data_Value_Alt'] = dataset1['Data_Value_Alt'].fillna(mean_Data_Value_Alt)
```

```python
dataset1.info()
```

```python
#filling missing values with most repeated values  
mode_Low_Confidence_Limit = dataset1['Low_Confidence_Limit'].mode()[0]
dataset1['Low_Confidence_Limit'] = dataset1['Low_Confidence_Limit'].fillna(mode_Low_Confidence_Limit)
```

```python
dataset1.info()
```

```python
mode_High_Confidence_Limit = dataset1['High_Confidence_Limit'].mode()[0]
dataset1['High_Confidence_Limit'] = dataset1['High_Confidence_Limit'].fillna(mode_High_Confidence_Limit)
```

```python
dataset1.info()
```

```python
#filling missing value with most repeated values in stratificationcategory2 attribute
mode_StratificationCategory2  = dataset1['StratificationCategory2'].mode()[0]
dataset1['StratificationCategory2'] = dataset1['StratificationCategory2'].fillna(mode_StratificationCategory2)

#filling missing value with most repeated values in stratification2 attribute
mode_Stratification2 = dataset1['Stratification2'].mode()[0]
dataset1['Stratification2'] = dataset1['Stratification2'].fillna(mode_Stratification2)

#filling missing value with most repeated values in geolocation attribute
mode_Geolocation = dataset1['Geolocation'].mode()[0]
dataset1['Geolocation'] = dataset1['Geolocation'].fillna(mode_Geolocation)
```

```python
# Visualization 1: Distribution of Cases by Year
plt.figure(figsize=(10, 6))
sns.countplot(x='YearStart', data=data, palette='viridis')
plt.title('Distribution of Cases by Year')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
# Find the year with the highest number of cases
year_with_max_cases = data['YearStart'].value_counts().idxmax()
year_with_max_cases
```

```python
# Filter data for the year with the highest number of cases
data_max_year = data[data['YearStart'] == year_with_max_cases]
```

```python
data_max_year['Stratification1'] = data_max_year['Stratification1'].astype('category')
```

```python
# Visualization 2: Distribution of Cases by Gender for the Year with the Highest Cases
plt.figure(figsize=(10, 6))
sns.histplot(data=data_max_year, x='Stratification1', palette='pastel', kde=False, discrete=True)
plt.title(f'Distribution of Cases by Gender in {year_with_max_cases}')
plt.xlabel('Gender')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
# Find the gender with the highest number of cases in that year
gender_with_max_cases = data_max_year['Stratification1'].value_counts().idxmax()
gender_with_max_cases
```

```python
# Filter data for the gender with the highest number of cases in that year
data_max_gender = data_max_year[data_max_year['Stratification1'] == gender_with_max_cases]
data_max_gender
```

```python
# Visualization 3: Distribution of Class IDs for the Gender with the Highest Cases
class_counts = data_max_gender['ClassID'].value_counts()

plt.figure(figsize=(12, 6))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Set2', len(class_counts)))
plt.title(f'Distribution of Class IDs for {gender_with_max_cases} in {year_with_max_cases}')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
class_id_counts = data['ClassID'].value_counts()
class_id_with_max_cases = class_id_counts.idxmax()
class_id_with_max_cases
```

```python
# Visualization 4: Distribution of Question IDs for the Most Frequent Class ID
data_max_class_id = data[data['ClassID'] == class_id_with_max_cases]
data_max_class_id
```

```python
topic_counts = data_max_class_id['TopicID'].value_counts().reset_index()
topic_counts.columns = ['TopicID', 'Count']
```

```python
# Visualization 4: Scatter Plot of Topic IDs for the Most Frequent Class ID
plt.figure(figsize=(14, 8))
plt.scatter(topic_counts['TopicID'], topic_counts['Count'], color='b')
plt.title(f'Distribution of Topic IDs for Class ID {class_id_with_max_cases}')
plt.xlabel('Topic ID')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 5: Distribution of High Confidence Limit and Low Confidence Limit for the Most Frequent Class ID
plt.figure(figsize=(14, 8))
sns.histplot(data_max_class_id, x='High_Confidence_Limit', color='blue', label='High Confidence Limit', kde=True, bins=30)
sns.histplot(data_max_class_id, x='Low_Confidence_Limit', color='red', label='Low Confidence Limit', kde=True, bins=30)
plt.title(f'Distribution of High and Low Confidence Limits for Class ID {class_id_with_max_cases}')
plt.xlabel('Confidence Limit')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
